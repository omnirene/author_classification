```python
prefeatures=author_dataframe['text']
prelabels=author_dataframe['author']# I called this prelabels because we need to process these values and convert them to numbers
```

```python
def tokenize_words(questions):
    tokens=[]
    for x in questions:
        tokens.append(nltk.word_tokenize(x))
    return tokens

def lowercase(features):
    new_features=[]
    for x in features:
        new_features.append([y.lower() for y in x])
    return new_features

def remove_unnecessary_characters(features):
    new_features=[]
    for x in features:
        new_features.append([re.sub(r'[^\w\s]','',y) for y in x])
    return new_features

def remove_stopwords(features):
    new_features=[]
    stopwords.words('english').append('one')#added this here after discovering it to be a common word between all authors
    for x in features:
        new_features.append([y for y in x if y not in stopwords.words('english')])
    return new_features

def lemmatize_words(features):
    lemmatizer=WordNetLemmatizer()
    lemmas=[]
    for x in features:
        lemmas.append([lemmatizer.lemmatize(y) for y in x])
    return lemmas

# def stem_words(features):
#     stemmer=LancasterStemmer()
#     stems=[]
#     for x in features:
#         stems.append([stemmer.stem(y) for y in x])
#     return stems

def prenormalize(questions):
    value=tokenize_words(questions)
    value=lowercase(value)
    value=remove_unnecessary_characters(value)
    value=remove_stopwords(value)
    value=lemmatize_words(value)
#     value=stem_words(value)
    return value
```

```python
prefeatures=prenormalize(prefeatures)
```
